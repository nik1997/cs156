```python
# get number of heads and get number of throws

import math
import pandas as pd

def nCr(n,r):
    f = math.factorial
    return f(n) // f(r) // f(n-r)


def em_algo(heads, throws, p_coinA = 0.6, p_coinB = 0.5):
    assert len(heads) == len(throws)
    
    test = 0
    
    while(True):
        
        #stores the probability of getting heads for each coin to check the convergence.
        last_p_coinA = p_coinA
        last_p_coinB = p_coinB
        
        test += 1
    
        arr_likelihood_A = []
        arr_likelihood_B = []

        arr_prob_A = []
        arr_prob_B = []

        arr_likely_num_heads_A = []
        arr_likely_num_tails_A = []
        arr_likely_num_heads_B = []
        arr_likely_num_tails_B = []

        for i in range(0, len(throws)):
            num_heads = heads[i]
            num_throws = throws[i]
            num_tails = num_throws - num_heads
            
            
            # E - step
            # Calculate likelihood of the throw being coinA or coinB
            likelihood_A = nCr(num_throws, num_heads) * (p_coinA**num_heads) * ((1 - p_coinA)**(num_throws - num_heads))
            likelihood_B = nCr(num_throws, num_heads) * (p_coinB**num_heads) * ((1 - p_coinB)**(num_throws - num_heads))

            arr_likelihood_A.append(likelihood_A)
            arr_likelihood_B.append(likelihood_B)

            # Normalizing the probabilities
            prob_A = likelihood_A / (likelihood_A + likelihood_B)
            prob_B = likelihood_B / (likelihood_A + likelihood_B)
            
            arr_prob_A.append(prob_A)
            arr_prob_B.append(prob_B)
            
            
            # M - step
            # Calculating likely number of heads for each coin
            likely_num_heads_A = prob_A * num_heads
            likely_num_tails_A = prob_A * num_tails
            arr_likely_num_heads_A.append(likely_num_heads_A)
            arr_likely_num_tails_A.append(likely_num_tails_A)


            likely_num_heads_B = prob_B * num_heads
            likely_num_tails_B = prob_B * num_tails
            arr_likely_num_heads_B.append(likely_num_heads_B)
            arr_likely_num_tails_B.append(likely_num_tails_B)


        p_coinA = sum(arr_likely_num_heads_A) / (sum(arr_likely_num_heads_A + arr_likely_num_tails_A))
        p_coinB = sum(arr_likely_num_heads_B) / (sum(arr_likely_num_heads_B + arr_likely_num_tails_B))

        print('New probabilities: Coin A = {0}; Coin B = {1}'.format(p_coinA, p_coinB))    

        # checking for convergence of the probability upto 2 decimal places.
        if round(p_coinA, 2) == round(last_p_coinA, 2) and round(p_coinB, 2) == round(last_p_coinB, 2):
            print('\n\n\nProbability of heads for: ')
            print('Coin A = {0:0.2f}'.format(p_coinA))
            print('Coin B = {0:0.2f}'.format(p_coinB))
            
            
            #Estimating the actual coin tossed in each throws
            arr_coin_toss = []
            for i in range(0, len(arr_prob_A)):
                if arr_prob_A[i] > arr_prob_B[i]:
                    arr_coin_toss.append('A')
                else:
                    arr_coin_toss.append('B')
            
            prob_table = pd.DataFrame(
                {'throws': heads,
                 'heads': throws,
                 'prob_coinA': arr_prob_A,
                 'prob_coinB': arr_prob_B,
                 'actual_coin_tossed': arr_coin_toss
                }, 
                columns=['throws', 'heads', 'prob_coinA', 'prob_coinB', 'actual_coin_tossed'])
            print(prob_table)
            
            break

    
    
    
# Starting Program
heads = [14, 33, 19, 10, 0, 17, 24, 17, 1, 36, 5, 6, 5, 13, 4, 35, 5, 5, 74, 34]
throws = [41, 43, 23, 23, 1, 23, 36, 37, 2, 131, 5, 29, 13, 47, 10, 58, 15, 14, 100, 113]
em_algo(heads, throws)
```
